```python
# pickle  loading
# we can later add more properties to each of these lists in hydro_dict
import pickle

with open('data.pickle', 'rb') as f:
    hydro_dict = pickle.load(f)
    
hydro_dict
```

```python
import numpy as np
import pandas as pd
```

```python
import math
```

```python
# find out the string with maximum length from train and test data

maxlen = 0

train_data = open('ACP500.txt','r')
lengths = list()

for line in train_data :
    if not line.startswith('>') :
        len_line = len(line)
        lengths.append(len_line)
        if len_line > maxlen :
            maxlen = len_line
            
test_data = open('ACP164.txt','r')

for line in test_data :
    if not line.startswith('>') :
        len_line = len(line)
        lengths.append(len_line)
        if len_line > maxlen :
            maxlen = len_line
            
lengths.sort()
percentage_to_average_or_max_over = 0.99
last_index_of_len_list_to_consider = math.ceil(len(lengths)*percentage_to_average_or_max_over) 
print(last_index_of_len_list_to_consider)
print(len(lengths)-1)
print(lengths[last_index_of_len_list_to_consider])
maxlen_to_consider = lengths[last_index_of_len_list_to_consider]
```

```python
# padding all the sequences to length 208(maximum sequence length in train and test combined) with Z

X_train_string_padded = list()
X_test_string_padded = list()
y_train = list()
y_test = list()

train_data = open('ACP500.txt','r')
train_labels = np.load('train_labels.npy')

c = -1

for line in train_data :
    if not line.startswith('>') :
        c = c + 1
        if not (len(line)) > maxlen_to_consider : 
            line = line.rstrip()
            y_train.append(train_labels[c])
            X_train_string_padded.append(line.ljust(maxlen_to_consider,'Z'))
        
test_data = open('ACP164.txt','r')
test_labels = np.load('test_labels.npy')

c = -1

for line in test_data :
    if not line.startswith('>') :
        c = c + 1
        if not (len(line)) > maxlen_to_consider :
            line = line.rstrip()
            y_test.append(test_labels[c])
            X_test_string_padded.append(line.ljust(maxlen_to_consider,'Z'))
        
```

```python
X_test_string_padded
```

```python
# number_of_channels will change according to number of properties in hydro_dict
X_train = list()
X_test = list()

number_of_channels = len(hydro_dict['A'])
sequence_length = maxlen_to_consider

for sequence in X_train_string_padded :
    
    each_instance = np.zeros((number_of_channels,sequence_length))
    for i,amino_acid in enumerate(sequence) :
        properties_of_the_amino_acid = hydro_dict[amino_acid]
        for j in range(number_of_channels) :
            each_instance[j,i] = properties_of_the_amino_acid[j]
            
    X_train.append(each_instance)
    
for sequence in X_test_string_padded :
    
    each_instance = np.zeros((number_of_channels,sequence_length))
    for i,amino_acid in enumerate(sequence) :
        properties_of_the_amino_acid = hydro_dict[amino_acid]
        for j in range(number_of_channels) :
            each_instance[j,i] = properties_of_the_amino_acid[j]
            
    X_test.append(each_instance)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)

print(X_train.shape)
print(X_test.shape)

np.save('X_train.npy',X_train)
np.save('X_test.npy',X_test)

np.save('train_labels.npy',y_train)
np.save('test_labels.npy',y_test)

X_train = np.load('X_train.npy')
y_train = np.load('train_labels.npy')
print(X_train.shape)
print(y_train.shape)
```
