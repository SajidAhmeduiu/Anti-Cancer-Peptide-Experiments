```python
from keras.layers import Dense
from keras.layers import Conv1D
from keras.layers import Flatten
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import MaxPool1D
from keras.callbacks import EarlyStopping
import numpy as np
import pandas as pd
from keras.layers import Dropout
from keras.layers import SpatialDropout1D
from keras.optimizers import Adam,SGD
from keras.layers import BatchNormalization
from keras.layers import GaussianDropout
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
from keras import backend as K
```

```python
X_train = np.load('X_train.npy')
y_train = np.load('train_labels.npy')

#shuffling the training instances
p = np.random.permutation(len(X_train))
X_train = X_train[p,:,:]
y_train = y_train[p]

#set the channels that you want to use
starting_channel = 0
ending_channel = 13
end_index = ending_channel + 1
number_of_channels_used = (ending_channel - starting_channel) + 1

X_train = X_train[:,starting_channel:end_index,:]

X_train_temp = X_train[0:448,:,:]
X_val = X_train[448:,:,:]
X_train = X_train_temp

y_train_temp = y_train[0:448]
y_val = y_train[448:]
y_train = y_train_temp




print(X_train.shape)
print(y_train.shape)
```

```python

visibile = Input(shape=(number_of_channels_used,94))
dropout_1 = SpatialDropout1D(0.3)(visibile)

conv_11 = Conv1D(filters=10,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(dropout_1)
#bn_11 = BatchNormalization()(conv_11)
#conv_12 = Conv1D(filters=30,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(conv_11)
#bn_12 = BatchNormalization()(conv_12)
spatial_dropout_1 = SpatialDropout1D(0.3)(conv_11)
max_pool_1 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_1)

conv_221 = Conv1D(filters=10,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(max_pool_1)
#bn_11 = BatchNormalization()(conv_11)
#conv_222 = Conv1D(filters=30,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(conv_221)
#bn_12 = BatchNormalization()(conv_12)
spatial_dropout_221 = SpatialDropout1D(0.3)(conv_221)
#max_pool_221 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first',padding='same')(spatial_dropout_221)

conv_21 = Conv1D(filters=10,kernel_size=6,padding='same',data_format='channels_first',activation='relu')(spatial_dropout_221)
#conv_21 = Conv1D(filters=30,kernel_size=6,padding='same',data_format='channels_first',activation='relu')(max_pool_1)
#bn_21 = BatchNormalization()(conv_21)
#conv_22 = Conv1D(filters=30,kernel_size=6,padding='same',data_format='channels_first',activation='relu')(conv_21)
#bn_22 = BatchNormalization()(conv_22)
spatial_dropout_2 = SpatialDropout1D(0.3)(conv_21)
max_pool_2 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_2)

conv_31 = Conv1D(filters=10,kernel_size=9,padding='same',data_format='channels_first',activation='relu')(max_pool_2)
#conv_31 = Conv1D(filters=30,kernel_size=9,padding='same',data_format='channels_first',activation='relu')(max_pool_2)
#conv_32 = Conv1D(filters=30,kernel_size=9,padding='same',data_format='channels_first',activation='relu')(conv_31)
spatial_dropout_3 = SpatialDropout1D(0.3)(conv_31)
max_pool_3 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_3)

conv_41 = Conv1D(filters=10,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(max_pool_3)
#conv_41 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(conv_32)
#conv_42 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(conv_41)
spatial_dropout_4 = SpatialDropout1D(0.3)(conv_41)
max_pool_4 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_4)

conv_51 = Conv1D(filters=10,kernel_size=15,padding='same',data_format='channels_first',activation='relu')(max_pool_4)
#conv_41 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(conv_32)
#conv_52 = Conv1D(filters=30,kernel_size=15,padding='same',data_format='channels_first',activation='relu')(conv_51)
spatial_dropout_5 = SpatialDropout1D(0.3)(conv_51)
max_pool_5 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_5)

#conv_61 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(max_pool_5)
#conv_41 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(conv_32)
#conv_62 = Conv1D(filters=30,kernel_size=12,padding='same',data_format='channels_first',activation='relu')(conv_61)
#spatial_dropout_6 = SpatialDropout1D(0.3)(conv_62)
#max_pool_6 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(spatial_dropout_6)

#conv_51 = Conv1D(filters=40,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(max_pool_4)
#conv_52 = Conv1D(filters=40,kernel_size=3,padding='same',data_format='channels_first',activation='relu')(conv_51)
#spatial_dropout_2 = SpatialDropout1D(0.5)(conv_22)
#max_pool_5 = MaxPool1D(pool_size=2,strides=2,data_format='channels_first')(conv_52)

flat_1 = Flatten()(max_pool_4)
#dropout_2 = Dropout(0.2)(flat_1)
#dropout_2 = GaussianDropout(0.2)(flat_1)

#dense_1 = Dense(40,activation='relu')(flat_1)
#bn_3 = BatchNormalization()(dense_1)
#dropout_3 = Dropout(0.2)(dense_1)
#dropout_3 = GaussianDropout(0.2)(dense_1)

dense_2 = Dense(20,activation='relu')(flat_1)
#bn_4 = BatchNormalization()(dense_2)
dropout_4 = Dropout(0.2)(dense_2)
#dropout_4 = GaussianDropout(0.2)(dense_2)

output = Dense(1,activation='sigmoid')(dropout_4)

model = Model(inputs=visibile, outputs=output)

print(model.summary())
```

```python
early_stopping_monitor = EarlyStopping(patience=1000)

```

```python
optimizer = Adam(lr=0.0005)
#optimizer = SGD(lr= 0.00001, momentum=0.9)

model_1 = model
model_1.compile(optimizer=optimizer, loss='binary_crossentropy',metrics=['accuracy'])
filepath = "model_21.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max',save_weights_only=False)
callback_list = [checkpoint]
model_1.fit(x=X_train,y=y_train,validation_data=(X_val,y_val), epochs=2000, callbacks=callback_list, batch_size=16)
```

```python
get_flatten_layer_output = K.function([model.layers[0].input],
                                  [model.layers[16].output])
```

```python
#transform train and test instances to the new space induces by cnn
## 1 cause we want to test, so dropout will be off
X_test = np.load('X_test.npy')
y_test = np.load('test_labels.npy')
X_train_tranformed = get_flatten_layer_output([X_train[:,:,:],1])[0]
X_test_tranformed = get_flatten_layer_output([X_test[:,:,:],1])[0]
print(X_train_tranformed.shape)
print(X_test_tranformed.shape)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = SVC(C=800000,kernel='rbf')
#clf = RandomForestClassifier() 

clf.fit(X_train_tranformed,y_train)
predictions = clf.predict(X_test_tranformed)

accuracy = accuracy_score(y_true=y_test,y_pred=predictions)

print(accuracy)
```
